```
import glob
import re
import os
import librosa
import random
import mix_package as mp
import pandas as pd 

#obtiene los paths para los gritos
paths_gr = glob.glob("ScreamsWAV/*.wav")
print('Número de gritos incluidos:',len(paths_gr))
```

```
#realiza la generación de samples
data = []
print('mixing...')
for i in range(10):
  try:
    os.makedirs('out/fold'+str(i+1))
  except:
    print('folder exists')
  paths_base = glob.glob("UrbanSound8K/audio/fold"+str(i+1)+"/*.wav")
  ij=0
  for pathi in paths_base:
    for pathj in paths_gr:
      final_path ='out/fold'+str(i+1)+'/'+str(ij)+'_'+str(i+1)+'.wav' 
      print(final_path,pathi,pathj)
      try:
        num = mp.sample_generator(final_path,pathi,pathj)
      except:
        print('exception occurs')
        num = -1
      if num!=-1:
        ij = ij+1
        print('mixed audio saved')
        data.append([final_path,pathi,pathj,i,num,1])
      else:
        print('too short base audio x.x')
      if ij==1000:
        break
    if ij==1000:
      break      
df = pd.DataFrame(data, columns=['path','path_base','path_scream','base_class','init_time','duration'])
```
